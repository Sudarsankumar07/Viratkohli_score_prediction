```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Score = np.arange(0,117,2)
dp = pd.read_excel('/content/Virat Kohli stats in IPL24.xlsx')
print('SCORES and Balls',Score)
```

```python
plt.plot(dp['Match Number'],dp['Score'], color='g')
plt.plot(dp['Match Number'],dp['Ball Faced'], color='b')
plt.scatter(dp['Match Number'],dp['Score'], color='r')
plt.scatter(dp['Match Number'],dp['Ball Faced'], color='b')
plt.xlabel('Match Number',fontsize=10)
plt.ylabel('Score and Balls')
plt.title('Runs Scored By Virat Kohli in IPL 24')
plt.xticks(fontsize=4.5)
for i in range(len(dp)):
  plt.text(dp['Match Number'][i], dp['Score'][i], dp['Opponent'][i], fontsize=10)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
a = model.fit(dp[['Ball Faced']],dp[['Score']])
```

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

```python
xfit = np.linspace(0, dp['Ball Faced'].max(), 100)
yfit = model.predict(xfit[:, np.newaxis])
plt.scatter(dp['Ball Faced'], dp['Score'])
plt.plot(xfit, yfit, color='red')
plt.xlabel('Ball Faced')
plt.ylabel('Score')
plt.title('Linear Regression of Score vs. Ball Faced')
plt.show()
```

```python
balls_faced = int(input("Enter the number of balls faced: "))
predicted_score = model.predict([[balls_faced]])
print("Predicted score:", predicted_score[0][0])
```
